# Graphs

```python
plt.style.use('dark_background')

for feature in features:
    plt.figure(figsize=(14, 10))
    sns.histplot(df[feature], kde=True)

    mean = df[feature].mean()
    median = df[feature].median()
    std = df[feature].std()
    plt.axvline(mean, color='red', linestyle='--', label='Mean')
    plt.axvline(median, color='green', linestyle='--', label='Median')
    plt.axvline(mean + std, color='black', linestyle='--', label='+1 Std')
    plt.axvline(mean - std, color='black', linestyle='--', label='-1 Std')
    plt.axvspan(mean - std, mean + std, color='yellow', alpha=0.2, label='+/- 1 Std Range')

    plt.title(f'{feature} Histogram with KDE')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

    print(f'Feature: {feature}')
    if df[feature].dtype in ['float64']:
        print('Type: Continuous')
    else:
        print('Type: Discrete')

    skewness = df[feature].skew()
    if skewness > 0:
        print('Skew: Right (Positive)')
    elif skewness < 0:
        print('Skew: Left (Negative)')
    else:
        print('Skew: None')

    kurtosis = df[feature].kurtosis()
    
    kurtosis += 3 # Adjusting for excess kurtosis
    
    if abs(kurtosis - 3.0) < 0.01:  # Consider values around 2.9 to be mesokurtic
        print('Kurtosis: Mesokurtic', kurtosis)
    elif kurtosis > 3:
        print('Kurtosis: Leptokurtic', kurtosis)
    elif kurtosis < 3:
        print('Kurtosis: Platykurtic', kurtosis)
    
        print('---')
```
